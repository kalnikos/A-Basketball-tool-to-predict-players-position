```python
## Theoretically the algorithm will be able to indentify the differences between the guards and centers but what about the forwards. I have 
## to find some unique features which will be able to help the algorithm
# Firstly lets examine the mean values of the features 

pos = df.groupby("Position").mean()

a = ['PER', 'TS%', '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%',
       'AST%', 'STL%', 'BLK%', 'TOV%', 'USG%', 'OWS', 'DWS', 'WS', 'WS/48',
       'OBPM', 'DBPM', 'BPM', 'VORP', 'FG', 'FG%', '3P', '3P%', '2P', '2P%',
       'eFG%', 'FT', 'FT%', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV',
       'PF',]

for j in a:
    values = [i for i, df in df.groupby("Position")]
    plt.bar(values, pos[j], color="b")

    plt.title("The mean value in "+ j + " Compared the three different positions")

    plt.xlabel("Positions")
    plt.ylabel("Values")

    plt.show()
```

```python
df = df2[['ORB%', 'DRB%', 'TRB%', 'AST%','BLK%','AST', 'STL', 'BLK', 'ORB',"3P","DBPM","3PAr","3P%", "Position"]]
## Correlation Matrix
plt.figure(figsize=(30,20))
corrMatrix = df.corr()
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
## Lets now examine the target group
df.Position.value_counts()
```

```python
## I will try to balance the observations 
from sklearn.utils import resample

df_guard = df[df.Position==1]
df_forward = df[df.Position==2]
df_Center = df[df.Position==3]

df_majority_downsampled = resample(df_guard, 
                                 replace=False,    # sample without replacement
                                 n_samples=len(df_forward),     # to match minority class
                                 random_state=123)

df_majority_downsampled2 = resample(df_Center, 
                                 replace=False,    # sample without replacement
                                 n_samples=len(df_forward),     # to match minority class
                                 random_state=123)
```

```python
df_downsampled = pd.concat([df_majority_downsampled,df_majority_downsampled2, df_forward])
df_downsampled.Position.value_counts()
```

```python
df_downsampled.shape
```

```python
## Saving the dataframe to a csv file in order to start the fun part
df_downsampled.to_csv("nba_predictions_data.csv")
```

```python
## It will be intresting to plot our data in 2 dimensions and indetify if there is a clear seperation. Let's use the PCA to plot our data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(df_downsampled)
```

```python
scaled_data = scaler.transform(df_downsampled)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
x_pca.shape
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = ['Guards', 'Forwards', 'Centers']
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = df_downsampled['Position'] == target
    plt.scatter(x_pca[:,0],x_pca[:,1],c=df_downsampled['Position'],cmap='rainbow')
ax.legend(targets)
ax.grid()
```
